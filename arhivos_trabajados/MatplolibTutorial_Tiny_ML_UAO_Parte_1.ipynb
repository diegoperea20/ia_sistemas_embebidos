##**Primer gráfico**

```
import matplotlib.pyplot as plt
plt.plot([1, 2, 4, 9, 5, 3])
plt.show()
```

* **plot** es la función encargada de contruir el grafico
*  **show** es la función que permite mostrar el gráfico

* la función **plot** recibe los datos en forma de vector




```
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.show()
```

El primer vector o parámetro de la función plot hace referencia al eje horizontal del gráfico. El segundo específica el eje vertical 

```
plt.plot([-3, -2, 5, 0], [1, 6, 4, 3])
plt.axis([-4, 6, 0, 7])
plt.show()
```

la funcion **axis** permite configurar el tamaño de los ejes del grafico

```
import numpy as np
x = np.linspace(-2, 2, 500)
y = x**2
plt.plot(x, y)
plt.show()
```

```
z= np.sin(x)
plt.plot(x,y,x,z)
plt.show()
```

A continuación se agregarán un título, los nombres de los y una cuadrícula.

```
plt.plot(x, y)
plt.title("Square function")
plt.xlabel("x")
plt.ylabel("y = x**2")
plt.grid(True)
plt.show()
```

# **Estilo y Corlor**

Por defecto, matlplotlib dibuja una línea entre puntos conseecutivos

```
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0])
plt.axis([-10, 110, -10, 140])
plt.show()
```

Existe un tercer argumento o parámetro, el cual indica el color y el estilo de la línea del gráfico

```
plt.plot([0, 100, 100, 0, 0, 100, 50, 0, 100], [0, 0, 100, 100, 0, 100, 130, 100, 0], "y:")
plt.axis([-10, 250, -10, 250])
plt.show()
```

Es posible graficar líneas con diferente estilo dentro de un mismo gráfico.

```
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-", [0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

Una forma mas ordenada se graficar varios datos en un mismo gráfico se muestra a continuación

```
plt.plot([0, 100, 100, 0, 0], [0, 0, 100, 100, 0], "r-")
plt.plot([0, 100, 50, 0, 100], [0, 100, 130, 100, 0], "g--")
plt.axis([-10, 110, -10, 140])
plt.show()
```

En lugar de dibujar líneas, es posible graficar los datos utilizando puntos con diferente forma y color

```
x = np.linspace(-6.28, 6.28, 30)
y=np.sin(x)
plt.plot(x, np.cos(x), 'g--', x, x, 'r:', x, y, 'b^')
plt.show()
```

Existes un gran número de atributos que se puede modificar para logar una apariencia deseada en el gráfico

```
x = np.linspace(-1.4, 1.4, 30)
line1, line2, line3 = plt.plot(x, x, 'g--', x, x**2, 'r:', x, x**3, 'b^')
line1.set_linewidth(3.0)
line1.set_dash_capstyle("round")
line3.set_alpha(0.2)
plt.show()
```

## **Guardar la figura**

La función savefig permite guardar la fígura en un formato de imagen que por lo general es png

```
x = np.linspace(-1.4, 1.4, 30)
plt.plot(x, x**2)
plt.savefig("my_square_function.png", transparent=True)
```
